# Signal representations

Notebook to look at different signal representations which allow classification of both time-locked and AM based signals

```python
import numpy as np
from sklearn.linear_model import Ridge, LogisticRegression
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

# evoked dataset

```python
N = 300
Y = np.random.randint(low=0,high=2,size=(N))
X = np.random.standard_normal((N,100))
sig = np.cos(np.linspace(0,2*np.pi*3,X.shape[1]))
X[Y==1,:] = X[Y==1,:] + sig[np.newaxis,:]
mu = np.vstack([np.mean(X[Y==i,:],0) for i in range(2)]).T
plt.imshow(np.vstack((X[Y==0,:],X[Y==1,:])),aspect='auto');plt.show()
plt.plot(mu)
```

```python
# classify:
clf = LogisticRegression()
scores = cross_val_score(clf,X,Y,cv=5)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
def filterbank(X,nbands=5,axis=-1):
    n = X.shape[axis]
    Fx = np.fft.fft(X,axis=axis)
    freqs = np.fft.fftfreq(n)
    bl = np.linspace(0,1/2,nbands+1)
    fb = []
    for bi in range(nbands):
        mask = np.logical_and(bl[bi] <= np.abs(freqs), np.abs(freqs) < bl[bi+1])
        Xb = np.fft.ifft(Fx*mask,axis=axis)
        print("{}-{} : {}".format(bl[bi],bl[bi+1],np.sum(np.abs(Xb.imag))))
        fb.append(Xb.real)
    X = np.stack(fb,axis=2)
    return X
```

```python
fbX = filterbank(X)
tmp = np.moveaxis(fbX,-1,0)
plt.imshow(tmp.reshape((-1,100)),aspect='auto')
```

```python
# classify:
clf = LogisticRegression()
scores = cross_val_score(clf,fbX.reshape((fbX.shape[0],-1)),Y,cv=5)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

# Induced

```python
# make an induced dataset
N = 300
Y = np.random.randint(low=0,high=2,size=(N))
X = np.random.standard_normal((N,100))
for i,l in enumerate(Y):
    if l==1:
        # random phase
        sig = np.cos(np.linspace(0,2*np.pi*3,X.shape[1]) + np.random.uniform(2*np.pi))
        X[i,:] = X[i,:] + sig
mu = np.vstack([np.mean(X[Y==i,:],0) for i in range(2)]).T
plt.imshow(np.vstack((X[Y==0,:],X[Y==1,:])),aspect='auto');plt.show()
plt.plot(mu)
```

```python
# classify:
clf = LogisticRegression()
scores = cross_val_score(clf,X,Y,cv=5)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
